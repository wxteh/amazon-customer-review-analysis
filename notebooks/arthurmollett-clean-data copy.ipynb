```python
def translate_batch(text_batch):
    translations = translator.translate(text_batch, dest="en")
    return [translation.text for translation in translations]
```

```python
batch_size = 100
```

```python
for i in range(0, len(review_content_clean), batch_size):
    batch = review_content_clean["review_content"].iloc[i:i+batch_size].tolist()
    translated_batch = translate_batch(batch)
    review_content_clean["review_content"].iloc[i:i+batch_size] = translated_batch
```
