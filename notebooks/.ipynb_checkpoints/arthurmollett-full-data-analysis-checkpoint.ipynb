```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, pipeline

# Specify the model name and revision
model_name = "facebook/bart-large-cnn"
revision = "main"  # or the specific revision you want to use

# Load the model and tokenizer
model = AutoModelForSeq2SeqLM.from_pretrained(model_name, revision=revision)
tokenizer = AutoTokenizer.from_pretrained(model_name, revision=revision)

summarizer = pipeline("summarization", model=model, tokenizer=tokenizer)

def get_summary(text, max_chunk_length=512, max_summary_length=25):
    # Split the text into chunks of max_chunk_length
    chunks = [text[i:i+max_chunk_length] for i in range(0, len(text), max_chunk_length)]

    summaries = []
    for chunk in chunks:
        summary = summarizer(chunk, max_length=max_summary_length, min_length=0, do_sample=False)
        summaries.append(summary[0]['summary_text'])

    # Combine the summaries of all chunks
    combined_summary = " ".join(summaries)
    return combined_summary

# grouping

grouped_df = cleaned_final_df.groupby("product_id")

summaries = []

for product_id, group in grouped_df:
    combined_reviews = " ".join(group['review_content'])
    summary = get_summary(combined_reviews)
    summaries.append({"product_id": product_id, "summary": summary})

# Create a DataFrame from the list of summaries
summary_df = pd.DataFrame(summaries)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
def sentiment_score(review):
    tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
    model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
    tokens = tokenizer.encode(review, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
from bs4 import BeautifulSoup
import requests
from transformers import pipeline, BertForSequenceClassification, BertTokenizer

# Initialize the summarization pipeline
summarizer = pipeline("summarization")

# Initialize the BERT sentiment analysis model and tokenizer
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)
sentiment_classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```
